```python
#딥러닝 구동하는데 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#필요한 라이브러리 불러옴
import numpy as np
import tensorflow as tf
```

```python
#실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분
np.random.seed(3)
tf.random.set_seed(3)
```

```python
#준비된 수술환자 데이터를 불러옴
Data_set = np.loadtxt('C:/Users/name_/Desktop/프로그래밍/딥러닝/deeplearning/dataset/ThoraricSurgery.csv', delimiter=',')
```

```python
#환자의 기록과 수술결과를 X와 Y로 저장
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

```python
#딥러닝 구조를 결정(모델을 설정하고 실행하는 부분)
model = Sequential() #딥러닝의 구조를 짜고 층을 설정하는 부분
model.add(Dense(30, input_dim=17, activation='relu')) #은닉층
model.add(Dense(1, activation='sigmoid')) #출력층
```

```python
#딥러닝이란 퍼셉트론 위에 숨겨진 퍼셉트론층 (은닉층)을 차곡차곡 추가하는 형태
#이 층들이 케라스에서는 Sequential() 함수를 통해 쉽게 구현된다
#Sequential 함수를 model로 선언하고 model.add()라는 라인을 추가하면 새로운 층이 만들어 진다
#맨 마지막 층은 출력층에 해당하고, 나머지는 모두 은닉층에 해당
#따라서 데이터에서 17개의 값을 받아 은닉층의 30개의 노드로 보낸다는 뜻
```

```python
#Dense()함수를 통해 현 은닉층에 몇개의 노드를 만들것인지 설정
#input_dim 변수는 입력데이터에서 몇개의 값을 가져올지를 설정
#케라스는 입력층을 따로 만드는 것이 아니라, 첫번째 은닉층에 input_dim을 추가해줌으로써 첫번째 Dense가 입력층+은닉층이 된다
#폐암수술환자의 생존여부데이터에는 17개의 입력데이터가 존재하므로 input_dim=17
```

```python
#은닉층의 각 노드는 17개의 입력값에서 임의의 가중치를 가지고 각 노드로 전송되어 활성화함수를 만난다 
#활성화함수를 거친 결괏값이 출력층에 전달된다
#이때 활성화 함수를 무엇으로 할지 지정하는 부분이 activation
#마지막층은 출력층이므로 노드가 1개
```

```python
#딥러닝 실행
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) #정해준 모델을 컴퓨터가 알아들을수 있게 컴파일
model.fit(X, Y, epochs=100, batch_size=10) #모델을 실제로 수행하는 부분
```

```python
#model.compile부분은 앞서 지정한 모델이 효과적으로 구현될 수 있게 여러가지 환경을 설정해 주면서 컴파일하는 부분
#loss = 'mean_squared_error' : 평균제곱오차함수를 오차함수로 사용
#metrics = ['accuracy'] : metrics는 모델이 컴파일될때 모델 수행 결과를 나타내게끔 설정하는 부분
```

```python
#컴파일단계에서 정해진 환경을 주어진 데이터를 불러 실행시킬때 사용되는 함수는 model.fit()
#폐암수술환자의 생존여부 데이터는 총 470명의 환자에게서 17개의 정보를 정리한 것
#이때 각 정보를 '속성'이라하고, 생존여부를 '클래스', 가로 한줄에 해당하는 각 환자의 정보를 각각 '샘플'이라고 한다
#주어진 데이터에는 총 470개의 샘플이 각각 17개의 속성을 가지고 있는 것이다
#학습 프로세스가 모든 샘플에 대해 한번 실행되는 것을 1 epoch라고 한다
#epoch=100으로 지정한 것은 각 샘플이 처음부터 끝까지 100번 재사용될 때까지 실행을 반복하라는 뜻
#batch_size는 샘플을 한번에 몇개씩 처리할지를 정하는 부분
#batch_size가 너무 크면 힉습속도가 느리고, 너무작으면 각 실행값의 편차가 생겨서 전체 결괏값이 불안정해 진다
```
