### 과적합 이해하기

```python
#과적합이란 모델이 학습 데이터셋 안에서는 일정수준 이상의 예측정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것을 말한다
#과적합은 층이 너무 많거나 변수가 복잡해서 발생하기도 하고 테스트셋과 학습셋이 중복될때 생기기도 한다
#특히 딥러닝은 학습 단계에서 입력층, 은닉층, 출력충의 노드들에 상당히 많은 변수들이 투입된다
#따라서 딥러닝을 진행할 동안 과적합에 빠지지 않게 늘 주의해야 한다
```

### 학습셋과 데이터셋

```python
#과적합을 방지하려면 학습을 하는 데이터셋과 이를 테스트하는 데이터셋을 완전히 구분한 다음 학습과 동시에 테스트를 병행해야 한다
#예를들어 데이터셋이 총100개의 샘플로 이루어져 있다면 70개는 학습셋으로, 30개는 테스트셋으로 사용한다
#신경망을 만들어 70개의 샘플로 학습을 진행한 후, 이 학습의 결과를 저장한다. 이렇게 저장된 파일을 모델이라 한다
#모델은 다른 셋에 적용할 경우 학습 단계에서 각인되었던 그대로 다시 수행한다
#따라서 나머지 30개의 샘플로 실험해서 정확도를 살펴보면 학습이 얼마나 잘 되었는지를 알 수 있다
#딥러닝같은 알고리즘을 충분히 조절하여 가장 나은 모델이 만들어지면 이를 실생활에 대입하여 활용하는 것이 바로 머신러닝의 개발 순서이다
```

```python
#지금까지는 테스트셋을 만들지 않고 학습을 했지만 매번 정확도를 계산할 수 있었다
#학습데이터를 이용해 정확도를 측정한 것은 데이터에 들어있는 모든 샘플을 그대로 테스트에 활용했기 때문이다
#이러한 방법은 빠른 시간에 모델 성능을 파악하고 수정할 수 있도록 도와준다
#하지만, 머신러닝의 최종목적은 과거의 데이터를 토대로 새로운 데이터를 예측하는 것이다
#즉, 새로운 데이터에 사용할 모델을 만드는 것이 최종 목적이므로 테스트셋을 만들어 정확한 평가를 병행하는것이 매우 중요하다
```

```python
#학습셋만 가지고 평가할때, 층을 더하거나 에포크값을 높여 실행 횟수를 늘리면 정확도가 계속해서 올라갈 수 있다
#하지만 학습데이터셋만으로 평가한 예측성공률이 테스트셋에서도 그대로 나타나지는 않는다
#즉, 학습이 깊어져서 학습셋 내부에서의 성공률은 높아져도 테스트셋에서는 효과가 없다면 과적합이 일어나고 있다는 것이다
#학습을 진행해도 테스트 결과가 더이상 좋지 않는 지점에서 학습을 멈춰야 한다 
#이때의 학습 정도가 가장 적절한 것으로 볼 수 있다
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
#seed값 설정
seed=0
np.random.seed(seed)
tf.random.set_seed(3)

#데이터 불러오기
df = pd.read_csv('C:/Users/name_/Desktop/프로그래밍/딥러닝/deeplearning/dataset/sonar.csv', header=None)
dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

### k겹 교차 검증

```python
#테스트를 더 정확하게 할수록 세상으로 나왔을 때 더 잘 작동한다
#하지만, 그만큼 데이터가 충분하지 않은 경우가 많이 있다
#딥러닝 혹은 머신러닝 작업을 할 때 늘 어려운 문제 중 하나는 알고리즘을 충분히 데스트하였어도
#데이터가 충분하지 않으면 좋은 결과를 내기 어렵다는 것이다
#이러한 단점을 보완하고자 만든것이 'k겹 교차검증'이다
#k겹 교차검증은 데이터셋을 여러개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법이다
#이렇게 하면 데이터의 100%를 테스트셋으로 사용할 수 있다
```

```python
from sklearn.model_selection import StratifiedKFold
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
#10개의 파일로 쪼개어 검증하는 10겹 교차 검증을 실시하도록 n_fold의 값을 10으로 설정한뒤 StratifiedKFold() 함수에 적용하였다
#그런다음 모델을 만들고 실행하는 부분을 for구문으로 묶어 n_fold만큼 반복되게 한다
```

```python
for train, test in skf.split(X,Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
```

```python
#또한 정확도를 매번 저장하여 한 번에 보여줄 수 있게 accuracy 배열을 만든다
accuracy = []

for train, test in skf.split(X,Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    k_accuracy = '%.4f' % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)

print('\n %.f fold accuracy : ' %n_fold, accuracy)
```
