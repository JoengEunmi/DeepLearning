### 데이터 분석과 입력

```python
#필요한 라이브러리 불러오기
import numpy as np
```

```python
#numpy는 수치계산을 위해 만들어진 라이브러리로 데이터 분석에 많이 사용
```

```python
#폐암 수술 환자의 수술 전 진단 데이터와 수술 후 생존 결과를 기록한 실제 의료 기록 데이터 불러오기
Data_set = np.loadtxt('C:/Users/name_/Desktop/프로그래밍/딥러닝/deeplearning/dataset/ThoraricSurgery.csv', delimiter =",")
```

```python
#환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
X = Data_set[:,0:17] #속성 데이터셋
Y = Data_set[:,17] #클래스 데이터셋
```

```python
#470명의 환자(470행)
#종양의 유형, 폐활량, 호흡곤란여부, 고통정도, 기침, 흡연, 천식여부 등 0~16가지 환자 상태(1~17번째 열)를 정리
#수술후 생존여부(18번째 열) : 1은 수술 후 생존, 0은 수술 후 사망
```

```python
#목적은 1번째부터 17번째까지의 항목을 분석해서 18번째 항목, 즉 수술 후 생존 또는 사망 여부를 맞추는 것
#1~17번째 항목을 속성(Attribute)라고 하고, 18번째 항목을 클래스(Class)라고 한다
```

```python
#딥러닝을 구동시키려면 '속성'만을 뽑아 데이터셋을 만들고, '클래스'를 담는 데이터셋을 또 따로 만들어 줘야한다
```

### 딥러닝 실행

```python
#딥러닝을 구동하는 데 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#모델을 설정하고 실행하는 부분, 딥러닝 구조를 결정
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Sequential() 함수는 딥러닝 구조를 한층 한층 쉽게 쌓아올릴 수 있게 한다
#Sequential() 함수를 선언하고 나서 model.add() 함수를 사용해 필요한 층을 추가해 준다
#keras의 가장 큰 장점은 model.add()를 통해 필요한 만큼의 층을 빠르고 쉽게 쌓아 올릴수 있다는 것
#Dense()함수는 각 층이 제각각 어떤 특성을 가질지 옵션을 설정하는 역할
```

```python
#activation = 다음 층으로 어떻게 넘길지 결정하는 부분, 가장 많이 사용하는 함수로는 relu()와 sigmoid() 함수를 사용
#loss = 한 번 신경망이 실행될 때마다 오차 값을 추적하는 함수
#optimizer = 오차를 어떻게 줄여 나갈지 정하는 함수
```

```python
#딥러닝 실행
model.compile(loss='mean_squared_error', optimizer='adam')
metrics=['accuracy']
model.fit(X, Y, epochs=100, batch_size=10)
```
