### 데이터 확인하기

```python
import pandas as pd
df = pd.read_csv('C:/Users/name_/Desktop/프로그래밍/딥러닝/deeplearning/dataset/housing.csv', delim_whitespace=True, header=None)
```

```python
#delim_whitespace=True : 공백을 기준으로 데이터를 나눈다
```

```python
df.info()
```

```python
import numpy as np
import tensorflow as tf

dataset = df.values
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
seed = 0
np.random.seed(3)
tf.random.set_seed(3)
```

### 선형회귀실행

```python
#선형회귀데이터는 마지막에 참과 거짓을 구분할 필요가 없다. 출력층에 활성화함수를 저정할 필요도 없다
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
#모델의 학습이 어느정도 되었는지 확인하기 위해 예측값과 실제값을 비교하는 부분을 추가한다
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, Y_test, Y_train = train_test_split(X,Y, test_size=0.3, random_state=seed)

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
Y_prediction = model.predict(X_test).flatten()
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print('실제가격 : {:.3f}, 예상가격{:.3f}'.format(label, prediction))
```

```python
#flatten() 함수는 데이터 배열이 몇 차원이든 모두 1차원으로 바꿔 읽기 쉽게 해주는 함수이다
```
